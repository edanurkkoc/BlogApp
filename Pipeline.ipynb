```python
from transformers import pipeline
```

```python
classifier=pipeline("sentiment-analysis")
```

```python
text="It's great to learn NLP for me"
```

```python
classifier(text)
```

```python
import pandas as pd
outputs=classifier(text)
pd.DataFrame(outputs)
```

```python
classifier=pipeline("zero-shot-classification")
```

```python
text="This is a tutarial about Hugging Face"
```

```python
labels=["tech","education","business"]
```

```python
outputs=classifier(text,labels)
pd.DataFrame(outputs)
```

```python
ner=pipeline("ner",grouped_entities=True)
```

```python
text="My name is Edanur from Turkey.HuggingFace is a nice platform"
```

```python
outputs=ner(text)
pd.DataFrame(outputs)
```

```python
reader=pipeline("question-answering")
```

```python
text="My name is Edanur and I love Turkey"
question="Where do I like"
```

```python
outputs=reader(question=question,context=text)
pd.DataFrame([outputs])
```

```python
summarizer=pipeline("summarization")
```

```python
text=""" Robert John Downey Jr. (born April 4, 1965)[1] is an American actor. His films as a leading actor have grossed over $14 billion worldwide, making him one of the highest-grossing actors of all time. Downey's career has been characterized by some early success, a period of drug-related problems and run-ins with the law, and a surge in popular and commercial success in the 2000s.[2] In 2008, Downey was named by Time magazine as one of the 100 most influential people in the world. From 2013 to 2015, he was listed by Forbes as Hollywood's highest-paid actor."""
```

```python
outputs=summarizer(text,max_length=60,clean_up_tokenization_spaces=True)
```

```python
print(outputs[0]["summary_text"])
```

```python
translator=pipeline("translation_en_to_de")
```
